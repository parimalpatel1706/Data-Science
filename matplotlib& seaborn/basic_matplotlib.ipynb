##### Barplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.bar(x,y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.barh(x,y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.bar(x,y,color='red')
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.bar(x,y,color='green')
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.bar(x,y,color='green', width=0.5)
plt.show()
```

### Pie 

```python
x =np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x =np.array([1,2,3,4])
mylabels=np.array(["apple","banana","mango","orange"])
plt.pie(x, labels=mylabels,startangle=90)
plt.show()
```

```python
x =np.array([1,2,3,4])
mylabels=np.array(["apple","banana","mango","orange"])
myexplode=np.array([0.3,0,0,0])
plt.pie(x, labels=mylabels,startangle=90,explode=myexplode)
plt.show()
```

```python
x =np.array([1,2,3,4])
mylabels=np.array(["apple","banana","mango","orange"])
myexplode=np.array([0.3,0,0,0])
plt.pie(x, labels=mylabels,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
x =np.array([1,2,3,4])
mylabels=np.array(["apple","banana","mango","orange"])
myexplode=np.array([0.3,0,0,0])
colour =np.array(["red","green","blue","yellow"])
plt.pie(x, labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=colour)
plt.show()
```

```python
x =np.array([1,2,3,4])
mylabels=np.array(["apple","banana","mango","orange"])
myexplode=np.array([0.3,0,0,0])
colour =np.array(["red","green","blue","yellow"])
plt.pie(x, labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=colour,autopct='%1.1f%%')
plt.show()
```

```python
x =np.array([1,2,3,4])
mylabels=np.array(["apple","banana","mango","orange"])
myexplode=np.array([0.3,0,0,0])
colour =np.array(["red","green","blue","yellow"])
plt.pie(x, labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=colour,autopct='%1.1f%%') 
plt.legend(title="Fruits")
plt.show()
```
