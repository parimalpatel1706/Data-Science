```python
x_point=np.array([1,3,4,5])
y_point=np.array([2,5,8,7])
plt.xlabel("x points")
plt.ylabel("y points")
plt.title("matplotlib study")
plt.grid()
plt.plot(x_point) 
plt.plot(y_point)
plt.show()
```

```python

colour=np.array(["red","green","blue","yellow","black"])# making an array of colour for point x and y
x1=np.array([1,2,3,4,5])
y1=np.array([1,8,27,64,125])
plt.scatter(x1,y1,c=colour)
plt.show()
```

```python
colour=np.array(["red","green","blue","yellow","black"])
x1=np.array([1,2,3,4,5])
y1=np.array([1,8,27,64,125])
plt.scatter(x1,y1,c=colour,alpha=0.5)# aplha used for increasing and decreasing colour apperance
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y,alpha=0.5, s = sizes, c = c )
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y,alpha=0.5, s = sizes, c = c ,cmap="coolwarm")
plt.colorbar()
plt.show()
```
