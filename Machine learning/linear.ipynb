```python
from sklearn.svm import SVC
svc_obj=SVC()
from sklearn.datasets import load_iris
x,y=load_iris(return_X_y=True)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
svc_obj.fit(x_train,y_train)
svc_obj.score(x_train,y_train)
svc_obj.score(x_test,y_test)
```

```python
svc_obj.predict(x_test)
```

```python
svc_obj.predict(y_test)
```

```python
## KNN classifier algorthim
from sklearn.neighbors import KNeighborsClassifier
knn_obj=KNeighborsClassifier()
knn_obj.fit(x_train,y_train)
knn_obj.score(x_test,y_test)
```

```python
knn_obj.score(x_train,y_train)
```

```python
## KNN regression
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
### Decision tree algorithm
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
dtr.score(x_test,y_test)
```
